```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
csv_path = Path('../data/main/master/weekly_Funda.csv')
csv = pd.read_csv(csv_path).set_index('nikkei')
PBR = csv['PBR']
```

```python
PBR_valid_length = len(PBR[PBR < 10])
print(PBR_valid_length)
PBR_nan_length = len(PBR[PBR.isna()])
print(PBR_nan_length)

error_length = len(PBR) - (PBR_valid_length + PBR_nan_length)
print(f'Number of erroneous PBR entries: {error_length}')
```

```python
sorted_csv = csv.sort_values('PBR', ascending=False).dropna()
```

```python
sorted_csv.to_csv('./tmp.csv')
```
