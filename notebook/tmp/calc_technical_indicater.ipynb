```python
from pathlib import Path
import pandas as pd
```

```python
tmp_path = Path("../data/nfq/nfqcsv")
tmp_files_set = set([f.stem for f in tmp_path.rglob('.') if f.suffix == '.csv'])
```

```python
len(tmp_files_set)
```

```python
check_sheet = pd.read_csv("../data/tickers/result/Tickers_with_Nikkei_copy.csv", )
nikkei = set(check_sheet["nikkei"].tolist())
ticker = set(check_sheet["ticker"].tolist())

print(len(check_sheet.index.values))
print(len(nikkei))
print(len(ticker))
```

```python
deleted_list = nikkei - tmp_files_set
mask = check_sheet["nikkei"].isin(deleted_list)
deleted_sheet = check_sheet[mask]
```
