## ラベルをさらに縮小する

```python
slim_df = pd.read_csv('../data/main/master/product/slim_weekly_FR.csv')
```

```python
origin_slim_df = slim_df.copy()
```

```python
cols = [
    '流動資産',
    '固定資産／非流動資産',
    '資産合計',
    '流動負債',
    '長期借入金',
    '資本金',
    '資本剰余金',
    '利益剰余金',
    '自己資本／親会社の所有者に帰属する資本',
    '売上総利益',
    '売上高・営業収益',
    '営業利益',
    '経常利益／税金等調整前当期純利益',
    '減価償却費',
    '法人税等',
    '営業活動によるキャッシュフロー',
    '投資活動によるキャッシュフロー',
    '現金及び現金同等物の期首残高',
    '現金及び現金同等物の期末残高',
    '現金及び預金',
]

def impute_critical_missing(df:pd.DataFrame, col:list[str]) -> pd.DataFrame:
    df = df.sort_values('week_id')
    for col in cols:
        df[col] = df[col].interpolate(method='linear', limit_direction='both')
    return df

origin_slim_df = (
    origin_slim_df
    .groupby('nikkei')
    .apply(lambda group: impute_critical_missing(df=group, col=cols))
    ).reset_index(drop=True)
```

```python
origin_slim_df.loc[origin_slim_df['株主資本'].isna() & origin_slim_df['上場フラグ'] == 1, '株主資本'] = (
    origin_slim_df['資本金'].fillna(0) +
    origin_slim_df['資本剰余金'].fillna(0) +
    origin_slim_df['利益剰余金'].fillna(0)
)

origin_slim_df.loc[origin_slim_df['株主資本'] == 0, '株主資本'] = np.nan
origin_slim_df.drop(columns=['資本金', '資本剰余金', '利益剰余金'], inplace=True)
print(f'Exit Ratio : {origin_slim_df.loc[origin_slim_df['上場フラグ'] == 1, '株主資本'].notnull().sum() / len(origin_slim_df.loc[origin_slim_df['上場フラグ'] == 1, '株主資本'])}')
```

```python
origin_slim_df.loc[origin_slim_df['税金等調整前当期純利益'].isna() & origin_slim_df['上場フラグ'] == 1, '税金等調整前当期純利益'] = (
    origin_slim_df['経常利益／税金等調整前当期純利益'].fillna(0) +
    origin_slim_df['特別利益'].fillna(0) -
    origin_slim_df['特別損失'].fillna(0)
)

origin_slim_df.loc[origin_slim_df['税金等調整前当期純利益'] == 0, '税金等調整前当期純利益'] = np.nan
print(f'Exit Ratio : {origin_slim_df.loc[origin_slim_df['上場フラグ'] == 1, '税金等調整前当期純利益'].notnull().sum() / len(origin_slim_df.loc[origin_slim_df['上場フラグ'] == 1, '税金等調整前当期純利益'])}')
```

```python
origin_slim_df['当期純利益'] = origin_slim_df['当期純利益（連結）']
origin_slim_df.drop(columns=['当期純利益（連結）'], inplace=True)
origin_slim_df.loc[origin_slim_df['当期純利益'].isna() & origin_slim_df['上場フラグ'] == 1, '当期純利益'] = (
    origin_slim_df['税金等調整前当期純利益'] -
    origin_slim_df['法人税等'].fillna(0)
)

origin_slim_df.loc[origin_slim_df['当期純利益'] == 0, '当期純利益'] = np.nan
print(f'Exit Ratio : {origin_slim_df.loc[origin_slim_df['上場フラグ'] == 1, '当期純利益'].notnull().sum() / len(origin_slim_df.loc[origin_slim_df['上場フラグ'] == 1, '当期純利益'])}')
```

```python
def group_function(df:pd.DataFrame, nikkei:str, cols2:str) -> pd.DataFrame:
    for col in cols2:
        print('\r', f'Processing nikkei: {nikkei}, column: {col}', end='               ')
        df[f'{col}の増加額'] = (
            df[col] - df[col].shift(1)
        )

        df.loc[df[f'{col}の増加額'] == 0, f'{col}の増加額'] = np.nan
        # print(f'Exit Ratio : {df.loc[df['上場フラグ'] == 1, f'{col}の増加額'].notnull().sum() / len(df.loc[df['上場フラグ'] == 1, f'{col}の増加額'])}')
```

```python
origin_slim_df_grouped = (
    origin_slim_df
    .groupby('nikkei')
    .apply
    (
        lambda group: group_function(df=group, nikkei=group.name, cols2=['売上総利益', '営業利益', '経常利益／税金等調整前当期純利益', '売上高・営業収益'])
    )
    .reset_index(drop=True))
```
