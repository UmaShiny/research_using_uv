```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
path = Path('../data/main/expansion/technical/daily_technical')
fp_list = [fp for fp in path.rglob('.') if fp.suffix == '.csv']
```

```python
print(len(fp_list))
```

```python
dropped_list:list[Path] = []
for fp in fp_list:
    print('\r', fp, end='')
    df = pd.read_csv(fp)
    df = df[df['nikkei'] == 'N0000001']
    if df.isnull().values.any():
        dropped_list.append(fp)

print('\nDropped rows found in the following files:')
for dropped in dropped_list:
    print(f' - {dropped.name}')
```

```python
ref_csv = pd.read_csv('../data/tickers/util_csv/nktk_cross_reference_rule.csv')
ref_ticker = ref_csv['ticker']

src_dir = Path('../data/main/build/jpx/jpx_dataset/fmt_4')
src_csvs = [fp for fp in src_dir.rglob('.') if fp.suffix == '.csv']

dst_dir = Path('../data/main/expansion/technical/extraction_empty_data')

for src_csv in src_csvs:
    base_df = ref_csv.copy()
    print('\r', src_csv, end='')
    df = pd.read_csv(src_csv)
    df.columns = ['date', 'ticker', 'jp', 'am_open', 'am_high', 'am_low', 'am_close', 'pm_open', 'pm_high', 'pm_low', 'pm_close']

    df['ticker'] = df['ticker'].astype(int) / 10
    df['ticker'] = df['ticker'].astype(int)

    mask = df['ticker'].isin(ref_ticker)
    masked_df = df[mask].drop(columns=['date', 'jp'])

    new_df = pd.merge(base_df, masked_df, on='ticker', how='left')

    new_df = new_df[~new_df.duplicated(subset=['nikkei'])]

    new_df.to_csv(dst_dir / src_csv.name, index=False)
```

```python
tmp_path = Path('../data/main/expansion/technical/extraction_empty_data')
src_csvs = [fp for fp in tmp_path.rglob('.') if fp.suffix == '.csv']

for src_csv in src_csvs:
    set_A = set(pd.read_csv(src_csv)['ticker'])
    set_B = set(ref_csv['ticker'])

    print(f'test_1: {set_B - set_A}')
    print(f'test_2: {set_A - set_B}')
    print('---')
```
