```python
finrep = Path('../data/main/build/nfq/nfqcsv_reduced_modify_quarterly/N0000001.csv')
df = pd.read_csv(finrep)
finrep_columns = pd.read_csv(finrep, nrows=0).columns.tolist()
```

```python
components = [col for col in finrep_columns]
```

```python
BS_SISAN = []
BS_HUSAI = []
BS_JYUNSISAN = []
PL = []
CF = []

target = BS_SISAN
for comp in components:
    if comp == '流動負債':
        target = BS_HUSAI
    elif comp == '純資産合計／資本合計':
        target = BS_JYUNSISAN
    elif comp == '売上高・営業収益':
        target = PL
    elif comp == '税金等調整前当期純利益':
        target = CF
    target.append(comp)
```

```python
finrep_columns_sorted = sorted(finrep_columns, key=lambda x: (len(x), x))

dict_col_include_name = {}
for col in finrep_columns_sorted:
    if '.1' in col:
        col = col[:-2]
    if col not in dict_col_include_name:
        dict_col_include_name[col] = []
    for target_col in finrep_columns_sorted:
        if col != target_col and re.search(re.escape(col), target_col):
            dict_col_include_name[col].append(target_col)

print(len(dict_col_include_name))
dict_col_include_name
```
