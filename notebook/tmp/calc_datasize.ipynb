```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
technical_data_path = Path('../data/technical')
all_technical_data_csv = [csv for csv in technical_data_path.rglob('.') if csv.suffix=='.csv']
print(f'Length of data: {len(all_technical_data_csv)}')
```

```python
cnt = 0
for csv in all_technical_data_csv:
    print('\r'f'processing: {csv.stem}, cnt={cnt}', end='')
    df = pd.read_csv(csv)
    cnt += len(df.index.values)
print('\ndone')
```
