```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
csv = Path('./weekly_remodify_fundamental_all.csv')
cols = pd.read_csv(csv, nrows=1).columns.tolist()
len(cols)
```

```python
missing_dict = {}
for col in cols:
    df = pd.read_csv(csv, usecols=[col])
    missing_ratio = df[col].isnull().sum() / len(df)
    missing_dict[col] = missing_ratio
    print(f'Column: {col}, Missing Ratio: {missing_ratio:.4f}')
```
