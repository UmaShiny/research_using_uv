```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
expand_tech_d = Path('../data/expansed_technical/daily_expansed_technical')
expand_tech_w = Path('../data/expansed_technical/weekly_expansed_technical')
expand_tech_m = Path('../data/expansed_technical/monthly_expansed_technical')
```

```python
d_total_missing = 0
for expand_tech_d_file in [f for f in expand_tech_d.rglob('.') if f.suffix == '.csv']:
    print('\r', f'{expand_tech_d_file}, missing={d_total_missing}', end='')
    daily_df = pd.read_csv(expand_tech_d_file, index_col=0)

    cols = daily_df.columns[3:7]  # 3〜7列目（0始まりなので注意）

    # 条件を分けて書く
    any_na = daily_df[cols].isna().any(axis=1)   # 1つでも欠損あり
    all_na = daily_df[cols].isna().all(axis=1)   # すべて欠損

    # 条件を組み合わせる
    mask = any_na & (~all_na)

    total_missing = mask.sum()
    d_total_missing += total_missing
print(f'daily total missing: {d_total_missing}')
```
