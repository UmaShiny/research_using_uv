```python
def detect_anomalies(group: pd.DataFrame):
    print('\r', f'Detecting anomalies for group: {group.name}', end=' ') 
    data = group[group['上場フラグ'] == 1].iloc[:, 2:]
    if len(data.values) == 0: 
        pred = np.nan
        score = np.nan
        print('\n', group.name, 'has no listed weeks.')
    else:
        data = data.fillna(0)
        clf = OneClassSVM(nu=0.05, kernel="rbf")
        clf.fit(data)
        pred = clf.predict(data)
        score = clf.decision_function(data)
    group.loc[data.index, 'anomaly'] = pred
    group.loc[data.index, 'anomaly_score'] = score
    return group
```

```python
BIG_df = pd.concat([weekly_FR_df_col42, FA_df.iloc[:, 2:], TR_df.iloc[:, 2:]], axis=1)
print(len(BIG_df))
```

```python
group_df = (
    BIG_df
    .groupby('nikkei')
    .apply(detect_anomalies)
).reset_index(drop=True)
```

```python
def group_filling_func(group: pd.DataFrame):
    print('\r', f'Filling missing values for group: {group.name}', end=' ')
    anomaly_mask = group['anomaly'] == -1
    group.loc[anomaly_mask, group.columns.to_list()[2:56]] = np.nan
    group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]] = group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]].interpolate()
    group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]] = group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]].fillna(group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]].mean())
    group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]] = group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]].fillna(0)
    if group.loc[group['上場フラグ'] == 1, group.columns.to_list()[2:56]].isnull().sum().sum() > 0:
        raise ValueError(f'{group.name} has NaN in {feature} after filling.')
    group = group.drop(columns=['anomaly', 'anomaly_score'])
    return group
```
