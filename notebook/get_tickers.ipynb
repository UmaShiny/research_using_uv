```python
import pandas as pd
import numpy as np
import os
import re
from pathlib import Path
from datetime import datetime as dt
```

### Tickerについて集計を行う

```python
path_to_csv_lst = sorted([p for p in Path(f"./../data/target_jpx_dataset").rglob(".") if p.suffix == ".csv"])
print(len(path_to_csv_lst))
```

```python
tickers_dict = {}
for path_to_csv in path_to_csv_lst:
    print("\r", path_to_csv, end="")
    df = pd.read_csv(path_to_csv)
    tickers = df.iloc[:, 1]
    date = path_to_csv.stem
    tickers_dict[date] = tickers
print("\nCompleted.")
```

```python
tickers_df = pd.DataFrame(tickers_dict)
tickers_df.to_csv("./../data/tickers/jpx_tickers.csv", index=False)
```

```python
tickers_df = pd.read_csv("./../data/tickers/jpx_tickers.csv")
all_tickers = tickers_df.values.flatten()
unique_tickers = np.unique(all_tickers)
print(f"Number of unique tickers: {len(unique_tickers)}")
```

```python
for i, ticker in enumerate(unique_tickers):
    print("\r", ticker, end="")
    if re.match(r'^[0-9]{4}.0$', str(ticker)) is None:
        print(f"\nInvalid ticker found: {i} >> {ticker}\n")
        # Remove invalid ticker
        unique_tickers = np.delete(unique_tickers, i)
print("\nCompleted checking tickers.")
print(f"Number of valid unique tickers: {len(unique_tickers)}")
```

```python
unique_tickers = unique_tickers.astype(int)
```

```python
pd.DataFrame(unique_tickers, columns=["Tickers"]).to_csv("./../data/tickers/unique_jpx_tickers.csv", index=False)
```

```python
ticker_name_pair_list = []
for path_to_csv in path_to_csv_lst:
    print("\r", path_to_csv, end="")
    df = pd.read_csv(path_to_csv)
    tickers = df.iloc[:, 1]
    name = df.iloc[:, 2]
    ticker_name_pair_list.extend(zip(tickers, name))
    ticker_name_pair_list = list(set(ticker_name_pair_list))
print("\nCompleted.")
print(f"Number of unique ticker-name pairs: {len(ticker_name_pair_list)}")
```

```python
ticker_name_pair_list.sort(key=lambda x: int(x[0]))
result = list({k: v for k, v in ticker_name_pair_list}.items())
print(f"Number of unique tickers after removing duplicates: {len(result)}")
```

```python
result_df = pd.DataFrame(result, columns=["Tickers", "Name"])
result_df["Tickers"] = result_df["Tickers"].astype(int)
result_df.to_csv("./../data/tickers/unique_jpx_ticker_name_pairs.csv", index=False)
```
