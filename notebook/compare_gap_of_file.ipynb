```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
funda_1 = Path("../data/nfq/concat_nfqxlsx/concat_stcvol_nfqxlsx/20000101-20191231")
funda_2 = Path("../data/nfq/nfqcsv/main_nfqcsv")
funda_3 = Path("../data/nfq/concat_nfqxlsx/concat_stcvol_nfqxlsx/20200101-20231231")
```

```python
funda_1_files = [f.stem for f in funda_1.rglob(".") if f.is_file()]
funda_2_files = [f.stem for f in funda_2.rglob(".") if f.is_file()]
funda_3_files = [f.stem for f in funda_3.rglob(".") if f.is_file()]
```

# concat 20000101-20191231 and 20200101-20231231

```python
in_set = set(funda_1_files) & set(funda_3_files)
print(len(in_set))
```

```python
for f in [f for f in all_set]:
    print('\r', f'processing : {f}', end=' ')
    if f in in_set:
        df = pd.concat([pd.read_excel(funda_1 / f'{f}.xlsx'), pd.read_excel(funda_3 / f'{f}.xlsx').iloc[:, 1:]], axis=1)
    else:
        if f in funda_1_files:
            df = pd.read_excel(funda_1 / f'{f}.xlsx')
        else:
            df = pd.read_excel(funda_3 / f'{f}.xlsx')

    # 保存前にA1だけをNaNにする
    cols = df.columns.tolist()
    cols[0] = np.nan  # 先頭列名をNaNに置き換え
    df.columns = cols

    df.to_excel(
        Path("../data/nfq/concat_nfqxlsx/concat_stcvol_nfqxlsx/20000101-20231231") / f'{f}.xlsx',
        index=False
    )
```
